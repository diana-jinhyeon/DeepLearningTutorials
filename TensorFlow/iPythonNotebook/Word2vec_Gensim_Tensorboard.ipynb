# word embedding (word2vec)
이번에는 word에 대한 representation중 하나인 word embedding을 다루겠습니다. <br>

#### Word embedding이란?
Word를 vector로 나타내는 기법을 의미합니다.<br>
word embedding은 glove, word2vec등 여러가지가 있지만 여기서는 제일 많이 쓰이는 word2vec에 대해서 다루겠습니다.<br>

#### 이전 one-hot representation의 한계
이전의 예제에서는 one-hot representation을 통해 Word를 vector화 했지만 one-hot representation은 몇가지 한계가 있습니다.<br>
- vector 간에 유사성을 계산할 수 없습니다 ( apple = [1, 0, ..., 0] 과 apples = [0, 1, 0, ... , 0] 은 완전히 다른단어로 표현됩니다 )
- 단어 하나를 나타낼때 전체 단어의 수 만큼의 차원을 할당하게 됨으로 효율 또한 낮습니다.


#### Word2vec으로 극복해보자
그렇다면, Word를 Vector로 바꿀 때, 어떻게 Vector에 수치를 정해주어야할까요? <br>
한 Word를 나타낼 수 있는 것은 무엇일까요? 핵심 아이디어는<br>
=> “Context” (문맥) !! 입니다. <br>
각 단어들은 context를 동반하고 context는 단어에 따라 다르며, 비슷한 단어들은 비슷한 context를 가집니다.<br>
이러한 아이디어로 단어를 벡터로 모델링 한 기법을 word2vec이라하며 크게 CBOW와 Skip-gram 두가지 방법이 많이 사용되고 있습니다.
![word2vec](./assets/skipgram.png)
#### How to word2vec
word2vec은 Tensorflow에서 제공하는 코드로 훈련시킬 수 도 있지만 대부분은 'gensim'이라는 모듈을 사용합니다.<br>
여기서도 'gensim'을 사용해서 word2vec을 training 해보고 embedding이 잘 되었는지 Tensorboard embedding projector를 통해 확인해보겠습니다.


```python
import nltk
import multiprocessing
from pprint import pprint
import time
# import codecs
# import re
import os
import numpy as np
import time
```

## Training에 사용할 Corpus - Reuters
Reuters Corpus는 <br>
-10,788 뉴스 문서 <br>
-1.3 million words<br>
-90개의 토픽<br>
으로 구성되어 있습니다.

```python
nltk.download('reuters')
from nltk.corpus import reuters
docs_en = [reuters.words(i) for i in reuters.fileids()] #reuters.fileids() == ['test/14826', 'test/14828', ... ,'training/9995']
print(docs_en)
```

```python
print("뉴스 문서 개수: ", len(docs_en))
print("Training에 사용할 보유 CPU 개수: ", multiprocessing.cpu_count())
```

```python
from gensim.models import word2vec
```

```python
config = {
    'min_count': 5,  # 등장 횟수가 5 이하인 단어는 무시
    'size': 50,  # 50차원짜리 벡터스페이스에 embedding
    'sg': 1,  # 0이면 CBOW, 1이면 skip-gram을 사용
    'batch_words': 1000,  # 사전을 구축할때 한번에 읽을 단어 수
    'iter': 2, #7,  # 보통 딥러닝에서 말하는 epoch과 비슷한, 반복 횟수를 의미 #너무 오래 걸릴땐 좀 낮춰야
    'workers': 1 #multiprocessing.cpu_count() #윈도우에서 에러
}
wv_model_en = word2vec.Word2Vec(**config)
wv_model_en.build_vocab(docs_en)
```

```python
count_t = time.time()
wv_model_en.train(docs_en) 
print('Running Time : %.02f' % (time.time() - count_t))
```

```python
wv_model_en.save('en_word2vec_50.model')
pprint(wv_model_en['I'])
pprint(wv_model_en.most_similar('I'))
pprint(wv_model_en['man'])
pprint(wv_model_en.most_similar('man'))
```

### 학습한 word2vec을 { word : Index } 의 Key : value 형태로 저장

```python
import json

vocab_path='./voca.json'
vocab = dict([(k, v.index) for k, v in wv_model_en.wv.vocab.items()])
with open(vocab_path, 'w') as f:
    f.write(json.dumps(vocab))
```

### 학습한 embedding을 2D Space에서 보기

```python
embedding_weights = wv_model_en.wv.syn0
final_embeddings = embedding_weights
labels = wv_model_en.wv.index2word
```

```python
%matplotlib inline
def plot_with_labels(low_dim_embs, labels, filename='tsne.png'):
  assert low_dim_embs.shape[0] >= len(labels), "More labels than embeddings"
  plt.figure(figsize=(18, 18))  # in inches
  for i, label in enumerate(labels):
    x, y = low_dim_embs[i, :]
    plt.scatter(x, y)
    plt.annotate(label,
                 xy=(x, y),
                 xytext=(5, 2),
                 textcoords='offset points',
                 ha='right',
                 va='bottom')

  plt.savefig(filename)

try:
  from sklearn.manifold import TSNE
  import matplotlib.pyplot as plt    


  tsne = TSNE(perplexity=30, n_components=2, init='pca', n_iter=5000)
  plot_only = 500
  low_dim_embs = tsne.fit_transform(final_embeddings[:plot_only, :])  
  labels = [labels[i] for i in range(plot_only)]
  plot_with_labels(low_dim_embs, labels)

except ImportError:
  print("Please install sklearn, matplotlib, and scipy to visualize embeddings.")
```

### 학습한 word2vec 파일 Loading

```python
word2vec_pretraining_file = './en_word2vec_50.model'
wv_model_en = word2vec.Word2Vec.load(word2vec_pretraining_file)

embedding_weights = wv_model_en.wv.syn0
final_embeddings = embedding_weights
labels = wv_model_en.wv.index2word
```

### Tensorboard의 embedding projector에서 학습한 embedding 확인

```python
import tensorflow as tf
from tensorflow.contrib.tensorboard.plugins import projector

embedding_var = tf.Variable(embedding_weights, name='embedding')

sess = tf.InteractiveSession()

tf.global_variables_initializer().run()
saver = tf.train.Saver()

# make log directory if not exists
log_dir = 'logs/'
if tf.gfile.Exists(log_dir):
    tf.gfile.DeleteRecursively(log_dir)
tf.gfile.MakeDirs(log_dir)
saver.save(sess, save_path=log_dir+'model.ckpt', global_step=None)

# TODO: add embedding data
summary_writer = tf.summary.FileWriter(logdir=log_dir, graph=tf.get_default_graph())
config = projector.ProjectorConfig()
embedding = config.embeddings.add()
embedding.tensor_name = embedding_var.name

# TODO: add label data
metadata_path = 'logs/metadata.tsv'
with open(metadata_path, 'w') as f:
    for i, label in enumerate(wv_model_en.wv.index2word):
        f.write('{}\n'.format(label))
embedding.metadata_path = metadata_path

# TODO: add sprite image (MNIST example)
# embedding.sprite.image_path = 'mnist/mnist_10k_sprite.png'
# embedding.sprite.single_image_dim.extend([28, 28])

# TODO: visualize embedding projector
projector.visualize_embeddings(summary_writer, config)
```
